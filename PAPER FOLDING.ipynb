```python
#PROBLEM 1
# WHEN PAPER IS FOLDED 43 times
THICKNESS = 0.00008
folded_thickness = THICKNESS *2**43
print("Thickness: {} meters".format(folded_thickness))
```

```python
#PROBLEM 2
#convert meters to kilometers and diplaying with two decimal places
print("Thickness: {: .2f} kilometers".format(folded_thickness / 10000))
```

```python
#PROBLME 3
#create using a for statement
thickness = 0.00008
folded_thickness = 0.00008*2
for i in range (1,43):
    folded_thickness = folded_thickness*2
print("Thickness: {: .2f} kilometers".format(folded_thickness / 10000))
```

```python
#PROBLEM 4
#Comparison of calculation time
import time
start = time.time()
THICKNESS = 0.00008
folded_thickness = THICKNESS *2**43
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
%%timeit
thickness = 0.00008
folded_thickness = 0.00008*2
for i in range (1,43):
    folded_thickness = folded_thickness*2
```

```python
#PROBLEM 5
#saving to a list
mylist = []
thickness = 0.00008
folded_thickness = thickness*2
mylist.append(thickness)
mylist.append(folded_thickness)
for i in range (1,43):
    folded_thickness = folded_thickness*2
    mylist.append(folded_thickness)
len(mylist)
```

```python
#PROBLEM 6
#Displaying a line graph
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(mylist)
plt.show()
#thickness in becomes thicker as the folds keeps on folding
```

```python
#PROBLEM 7
#customizing graphs
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(mylist, color='green', linestyle='dotted', linewidth=3)
plt.show()
```

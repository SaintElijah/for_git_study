```python
#PROBLEM 1
import numpy as np
mean = [-3,0]
covariance = [[1.0,0.8],[0.8,1.0]]
seed = np.random.seed(3)
rand1 = np.random.multivariate_normal(mean,covariance, 500)
print(rand1)
```

```python
rand1.shape
```

```python
#PROBLEM 2
from matplotlib import pyplot as plt
%matplotlib inline
plt.title("Scatter plot of the random values")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand1[:,0],rand1[:,1])
```

```python
#PROBLEM 3
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("Histogram")
plt.hist(rand1[:,0])
plt.xlim(left=-3)
plt.show()
```

```python
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("Histogram")
plt.hist(rand1[:,1])
plt.xlim(right=2)
plt.show()
```

```python
#PROBLEM 4
mean = [0,-3]
covariance = [[1.0,0.8],[0.8,1.0]]
seed = np.random.seed(3)
rand2 = np.random.multivariate_normal(mean,covariance, 500)
print(rand2)
```

```python
#plotting graphs for problem 4
plt.title("Two scatter plot in one chart")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand1[:,0], rand1[:,1], s=20, label = "0", marker="v")
plt.scatter(rand2[:,0], rand2[:,1], s=20, label = "1", marker="o")
plt.legend()
plt.show()
```

```python
#PROBLEM 5
combine = np.concatenate((rand1,rand2))
print(combine)
```

```python
combine.shape
```

```python
#PROBLEM 6
zeros = np.zeros(500)
ones = np.ones(500)
zeros_ones = np.concatenate((zeros,ones))
ndarr = np.append(combine,zeros_ones[:,None],1)
print(ndarr)
```

```python
def time_covering(smaller_size, targeted_size):
    count = 0
    minutes = 0
    while smaller_size <= targeted_size:
        smaller_size = smaller_size*2
        count+=1
        minutes = count*5
    return minutes

#let say the size of the chestnut is around 6g and the solar system is around 6.95700Ã—10**8, according to google
#the time for the chestnut to reach the solar system after the spinkled would be
print(time_covering(6,69570000000000),"minutes")
```

```python
#time it would take the tokyo dome to fill up
#according to google the tokyo dome is of size 46,755m square and football size is 22cm
tokyo_dome_size = 46755
ball_size = 22/100
print(time_covering(ball_size,tokyo_dome_size),"minutes")
```

```python
#let plot a line graph for the example of the tokyo dome
#let create a list
sizes = []
minutes = []
def time_covering_list(smaller_size, targeted_size):
    count = 0
    minute = 0
    sizes = []
    minutes = []
    while smaller_size <= targeted_size:
        sizes.append(smaller_size)
        minutes.append(minute)
        smaller_size = smaller_size*2
        count+=1
        minute = count*5
    return minutes,sizes
x,y = time_covering_list(ball_size,tokyo_dome_size)
print(x)
print(y)
```

```python
#Displaying a line graph
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("time taken for ball to fill up tokyo dome")
plt.xlabel("number of minutes")
plt.ylabel("number of size")
plt.plot(x,y)
plt.show()
```

```python
#get nth in list
grain = 1
grainlist = [grain,]
for i in range (1,100):
    grain = grain*2
    grainlist.append(grain)
#len(grainlist)
```

```python
#Displaying a line graph
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.title("total number of rice receive from 1-100 days")
plt.xlabel("number of days")
plt.ylabel("number of rice")
x = np.arange(100)
y = grainlist
plt.plot(y,x)
plt.show()
```

```python
#PROBLEM 3: How many people can live for how many days with the rice we get
def myfunction (n_grains,n_people):
    
    #according to stlukeshealth.org, average grains per day should be 6
    GRAINS_PER_DAY = 6
    
    #number grains required for number of people, explained as total grains needed to people
    total_grains_require = n_people*GRAINS_PER_DAY
    
    #now calculate the number of days this people will survive
    #but first let check the number of grains is sufficent for the people per day
    if n_grains >=total_grains_require:
        n_days_live = n_grains/total_grains_require #number of days to live
        return n_days_live
    else:
        return "The number of grains can not survive the people in a day"
#now let output a valid argument, say 500000grains for 20people
myfunction(500000,20)
```
